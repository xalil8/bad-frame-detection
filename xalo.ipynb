```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 

```

```python
bad = cv2.imread("deneme_1.png")
img2 = cv2.imread("deneme_2.png")
img3 = cv2.imread("deneme_3.png")

y,x = bad.shape[:2]
```

```python
badimg = cv2.cvtColor(bad, cv2.COLOR_RGB2GRAY)
im2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
im3 = cv2.cvtColor(img3, cv2.COLOR_RGB2GRAY)

```

```python
bad_val = cv2.countNonZero(badimg)
val2 = cv2.countNonZero(im2)
val3 = cv2.countNonZero(im3)
print(bad_val,val2,val3)
```

```python
diff1 = abs(bad_val-val2)
diff2 = abs(val2-val3)
print(diff1,diff2)
```

```python
sum1 = cv2.sumElems(bad)
sum2 = cv2.sumElems(img2)
sum3 = cv2.sumElems(img3)


diff1 = abs(sum(sum1) - sum(sum2))
diff2 = abs(sum(sum2) - sum(sum3))

print(diff1/(x*y), diff2/(x*y))

```

```python
sum1 = cv2.sumElems(bad)
sum2 = cv2.sumElems(img2)
sum3 = cv2.sumElems(img3)


diff1 = abs(sum(sum1) - sum(sum2))
diff2 = abs(sum(sum2) - sum(sum3))

print(diff1/(x*y), diff2/(x*y))

```
